```python
import pandas as pd
from math import sqrt
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
d = pd.read_csv("/Users/pouya/Machine Learning Course/diamonds.csv",index_col=0)
d.shape
```

```python
d.head(7)
```

```python
d.color= d.color.apply(list('JIHGFED').index)
d.head(7)
```

```python
d.cut= d.cut.apply(['Fair','Good','Very Good','Premium','Ideal'].index)
d.head(7)
```

```python
d.clarity= d.clarity.apply(['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF'].index)
d.head(7)
```

```python
d.describe()
```

```python
x= d.loc[:,(d.columns!= 'price')]
y = d.price
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=.30,random_state=1)
xtrain.shape
```

```python
model= LinearRegression()
newxtrain= xtrain.loc[:,(xtrain.columns !='cut')& (xtrain.columns!='color')&(xtrain.columns!='clarity')]
newxtest= xtest.loc[:,(xtest.columns!='cut')& (xtest.columns!='color')&(xtest.columns!='clarity')]
model.fit(newxtrain,ytrain)
ypred= model.predict(newxtest)
```

```python
ypred.shape
```

```python
mean_absolute_error(ytest,ypred)
```

```python
sqrt(mean_squared_error(ytest,ypred))
```
