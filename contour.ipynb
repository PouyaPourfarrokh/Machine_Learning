```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
x= np.linspace(0,1,num=500)
y= np.linspace(0,1,num=500)
```

```python
X,Y= np.meshgrid(x,y)
Z= X+Y
```

```python
fig= plt.figure(figsize=(5,5))
ax= fig.gca(projection='3d')
s= ax.plot_surface(X,Y,Z, cmap=plt.cm.rainbow)
cset= ax.contour(X,Y,Z, zdir='z',offset=0, cmap=plt.cm.rainbow)
fig.colorbar(s,shrink=0.5,aspect=5);
```

```python
x=np.arange(-10,10,0.5)
y=np.arange(-10,10,0.5)
X,Y= np.meshgrid(x,y)
Z=X**3+ Y**4
```

```python
fig= plt.figure(figsize=(5,5))
ax= fig.gca(projection='3d')
s= ax.plot_surface(X,Y,Z, cmap=plt.cm.coolwarm)
cset= ax.contour(X,Y,Z, zdir='z',offset=0, cmap=plt.cm.coolwarm)
fig.colorbar(s,shrink=0.5,aspect=5);
```

```python
x=np.arange(1,11)
y=x.reshape(-1,1)
h= x*y+np.sin(x)
h
```

```python
cs= plt.contour(h,levels=[10,30,40], colors=['red','blue'], extend='both')
cs.cmap.set_over('yellow')
cs.cmap.set_under('pink')
cs.changed()
```
